```python
##Data Transform##
#Drop Rows with missing count
```

```python
df_dropped = df.dropna(subset=['Self_Employed'])
df_dropped.head()
```

```python
df_dropped.shape
```

```python
df_copy = df.copy()
df_copy.head()
```

```python
df_copy.shape
```

```python
#Fill Missing count values with mean count

#ffill() and bfill() functions copy forward the data from the previous row (forward fill) or copy the value from the next row (backward fill)
```

```python
df_dropped['LoanAmount'].dtype
```

```python
df_dropped['LoanAmount'].mean()
```

```python
df_copy['LoanAmount'].fillna(value=np.round(df['LoanAmount'].mean(),decimals=0), inplace=True)
```

```python
df_copy.agg(['count', 'size', 'nunique'])
```

```python
##Handling Categorical Data
```

```python

list(df['Gender'].unique())
```

```python

"""
1 Male
2 Female
3 nan"""
df['Gender'].replace({"Male":1, 
                       "Female":2, 
                       "nan":3}, inplace=True)
```

```python
list(df['Gender'].unique())
```

```python
##Create a dictionary
```

```python
keys = list(df['Gender'].unique())
vals = range(1,8)
act = dict(zip(keys, vals))
act
```

```python

df_copy['Gender'] = df['Gender'].map(act)
df_copy.head()
```
